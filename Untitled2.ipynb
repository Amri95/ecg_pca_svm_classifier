```python
AMP = 0.2
N = 10000  # number of samples
k = 120000
qN = 0.02  # quantization stepsize

def uniform_midtread_quantizer(x, Q):
    # limiter
    x = np.copy(x)
    idx = np.where(np.abs(x) >= 1)
    x[idx] = np.sign(x[idx])
    # linear uniform quantization
    xQ = qN * np.floor(x/qN + 0.5)

    return xQ

def plot_signals(x, xQ):
    e = xQ-x
    plt.figure(figsize=(8,8))
    plt.plot(x, label='fx')
    plt.plot(xQ, label='QN')
    plt.plot(e, label='D(QN)')
    plt.xlabel('k')
    plt.axis([0, N,  -0.1, 0.5])
    plt.legend()
    plt.grid()
```

```python
# generate signal
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 10000)
fx1 = norm.pdf(x)
# quantize signal
QN = uniform_midtread_quantizer(fx1, Q)
# plot signals
plot_signals(fx1, QN)

H_f1 = entropy(vv1.pdf(x)*vv1.logpdf(x))
H_QN = entropy(QN)
print(H_f1)
print(H_QN)

print(np.log(np.sqrt(12* sum(pow(fx1-QN,2)*fx1))))
pow(qN,2)/12
```
